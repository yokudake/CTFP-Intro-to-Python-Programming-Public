<h3> Preview of the Next Tutorial: Plotting Probability Densities in more than one dimention </h3>

In the next tutorial, we'll learn how to use functional programming along with matplotlib to plot probability densities that have 2 or 3 dimentions. These include particle in a 2D box, and atomic orbitals! See below for an example of what we'll tackle in the next tutorial.

#### Run the code below to see a plot a 2D and 3D representation of the equation $z = x^{2} + y^{2}$

```python
import numpy as np #importing the numpy module
import matplotlib.pyplot as plt #importing the matplotlib.pyplot module


def f(x, y):
    # returns the value of x^2 + y^2
    return x**2 + y**2

x_axis = np.linspace(start = -10, stop = 10, num = 21) #create an array for the x axis
y_axis = np.linspace(start = -10, stop = 10, num = 21) #create an array for the y axis

x, y = np.meshgrid(x_axis, y_axis) #creates a 2D array for the x and y axes
z = f(x, y) #defines z as the output of the function f(x,y)

## Plotting in 2D ##
#calling the method plt.contour() to plot a 2D representation of the equation x^2 + y^2
contour_plot_1 = plt.contourf(x, y, z, cmap = "bone") 

#calling the method plt.colorbar() to display a label for the values of z shown in contour_plot_1
plt.colorbar(contour_plot_1, label = "z axis") 

#calling the method plt.xlabel() to display a label for the xaxis
plt.xlabel("x axis")

#calling the method plt.ylabel() to display a label for the yaxis
plt.ylabel("y axis")

#calling the method plt.show() to plot the current axis as its own figure
plt.show()

## Plotting in 3D ##
#defining the axis as one that will be plotted in 3D
ax = plt.axes(projection = "3d")

#calling the method ax.plot_surface to plot a 3D representation of the equation x^2 + y^2
ax.plot_surface(x, y, z, cmap = "bone")

#calling the method ax.set_xlabel() to display a label for the xaxis
ax.set_xlabel("x axis")

#calling the method ax.set_ylabel() to display a label for the xaxis
ax.set_ylabel("y axis")

#calling the method ax.set_zlabel() to display a label for the xaxis
ax.set_zlabel("z axis")

#calling the method plt.show() to plot the current axis as its own figure
plt.show()
```
