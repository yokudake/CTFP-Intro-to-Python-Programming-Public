Recall that a solution for the particle in a 1D box is given by the equation:

$\large \psi(x, n, a) = \sin(\frac{n\pi x}{a})$ 

for $\large 0 \leq x \leq a$ where $n = 1,2,3...$

Where a solution to a particle in a 2D box is dependent on quantum numbers $n_{x}$ and $n_{y}$ and two *position axes*
$x$ and $y$:

$\large \psi(x, y, n_{x}, n_{y}, a,b) = \left(\sin{\frac{n_{x}\pi x}{a}}\right)
\left(\sin{\frac{n_{y}\pi y}{b}}\right)$ 

for $\large 0 \leq x \leq a$ and $\large 0 \leq y \leq b$ where $n = 1,2,3...$

We will walk through how to plot this solution to particle in 2D box and use it as an example of how to make contour plots in **matplotlib**.
